### Extracting Location Info and Sorting by Location

```python
def extract_unit(locator):
    locator = str(locator)
    locator = locator.split('-')
    locator = locator[5].strip()
    return int(locator)

inventory.insert(0, "Unit", inventory['LocatorCode'].apply(extract_unit), True)
```

```python
def extract_zone(locator):
    locator = str(locator)
    locator = locator.split('-')
    return locator[1].strip()

def extract_aisle(locator):
    locator = str(locator)
    locator = locator.split('-')
    return locator[2].strip()

def extract_bay(locator):
    locator = str(locator)
    locator = locator.split('-')
    locator = locator[3].strip()
    return int(locator)

inventory['Zone'] = inventory['LocatorCode'].apply(extract_zone)
inventory['Aisle'] = inventory['LocatorCode'].apply(extract_aisle)
inventory['Bay'] = inventory['LocatorCode'].apply(extract_bay)
```

```python
inventory.sort_values(['Zone', 'Aisle', 'Bay', 'Unit'], ascending=[True, True, True, True], inplace=True)
```

**Adds Zone, Aisle, and Bay Rows to the Dataframe**

```python
for index, row in inventory.iterrows():
    length = len(inventory)
    if inventory['Zone'].iloc[index] != inventory['Zone'].iloc[index - 1]:
        row_input = pd.Series({'Unit':'Zone ' + inventory['Zone'].iloc[index], 'LocatorCode':' ', 'Q':' ', 'Title':' ', 'Condition':' ', 'Sales Rank':' ', 'Price':' ', 'AZ Low':' ', 'Received':' '} , name=(index-0.3))
        inventory_print = inventory_print.append(row_input)
    if inventory['Aisle'].iloc[index] != inventory['Aisle'].iloc[index - 1]:
        row_input = pd.Series({'Unit':'Aisle ' + inventory['Aisle'].iloc[index], 'LocatorCode':' ', 'Q':' ', 'Title':' ', 'Condition':' ', 'Sales Rank':' ', 'Price':' ', 'AZ Low':' ', 'Received':' '} , name=(index-0.2))
        inventory_print = inventory_print.append(row_input)
    if inventory['Bay'].iloc[index] != inventory['Bay'].iloc[index - 1]:
        row_input = pd.Series({'Unit':'Bay ' + str(inventory['Bay'].iloc[index]), 'LocatorCode':' ', 'Q':' ', 'Title':' ', 'Condition':' ', 'Sales Rank':' ', 'Price':' ', 'AZ Low':' ', 'Received':' '} , name=(index-0.1))
        inventory_print = inventory_print.append(row_input)
    index += 1
```
